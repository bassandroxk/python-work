#### 1. Create a class of student (name, sap id, marks[phy,chem,maths]). Create 3 objects by taking inputs from the user and display details of all students. 

```python

class Student:
    def __init__(self, name, sap_id, marks):
        self.name = name
        self.sap_id = sap_id
        self.marks = marks  

students = [
    Student("Alice", "12345", [85, 90, 88]),
    Student("Bob", "12346", [75, 80, 70]),
    Student("Charlie", "12347", [95, 85, 92])
]

print("Task 1: Student Details:")
for student in students:
    print(f"Name: {student.name}, SAP ID: {student.sap_id}, Marks: {student.marks}")
print()
```

<h4>2. Add constructor in the above class to initialize student details of n students and implement following methods:  <br>
Display() student details  <br>
Find Marks_percentage() of each student  <br>
Display result() [Note: if marks in each subject >40% than Pass else Fail]  Write a Function to find average of the class.  </h4>

```python

class Student:
    def __init__(self, name, sap_id, marks):
        self.name = name
        self.sap_id = sap_id
        self.marks = marks  
    def display(self):
        print(f"Name: {self.name}, SAP ID: {self.sap_id}, Marks: {self.marks}")

    def marks_percentage(self):
        total = sum(self.marks)
        return (total / 300) * 100  

    def display_result(self):
        percentage = self.marks_percentage()
        result = "Pass" if all(mark >= 40 for mark in self.marks) else "Fail"
        print(f"Result: {result} (Percentage: {percentage:.2f}%)")

students = [
    Student("Alice", "12345", [85, 90, 88]),
    Student("Bob", "12346", [75, 80, 70]),
    Student("Charlie", "12347", [95, 85, 92])
]

print("Task 2: Enhanced Student Class:")
for student in students:
    student.display()
    print(f"Percentage: {student.marks_percentage():.2f}%")
    student.display_result()
    print()


def class_average(students):
    if not students:
        return 0
    total_percentage = sum(student.marks_percentage() for student in students)
    return total_percentage / len(students)

print(f"Class Average Percentage: {class_average(students):.2f}%")
print()
```

#### 3. Create programs to implement different types of inheritances.  

```python

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

print("Task 3a: Single Inheritance:")
dog = Dog()
dog.speak()
dog.bark()

class Father:
    def father_trait(self):
        print("Father's trait")

class Mother:
    def mother_trait(self):
        print("Mother's trait")

class Child(Father, Mother):
    def child_trait(self):
        print("Child's trait")

print("\nTask 3b: Multiple Inheritance:")
child = Child()
child.father_trait()
child.mother_trait()
child.child_trait()

class Grandparent:
    def grandparent_method(self):
        print("Grandparent method")

class Parent(Grandparent):
    def parent_method(self):
        print("Parent method")

class Child2(Parent):
    def child_method(self):
        print("Child method")

print("\nTask 3c: Multilevel Inheritance:")
child2 = Child2()
child2.grandparent_method()
child2.parent_method()
child2.child_method()

class Vehicle:
    def vehicle_info(self):
        print("This is a vehicle")

class Car(Vehicle):
    def car_info(self):
        print("This is a car")

class Bike(Vehicle):
    def bike_info(self):
        print("This is a bike")

print("\nTask 3d: Hierarchical Inheritance:")
car = Car()
car.vehicle_info()
car.car_info()

bike = Bike()
bike.vehicle_info()
bike.bike_info()

class A:
    def method_a(self):
        print("Method A")

class B(A):
    def method_b(self):
        print("Method B")

class C(A):
    def method_c(self):
        print("Method C")

class D(B, C):
    def method_d(self):
        print("Method D")

print("\nTask 3e: Hybrid Inheritance:")
d = D()
d.method_a()
d.method_b()
d.method_c()
d.method_d()
print()
```

#### 4. Create a class to implement method Overriding.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):  
        print("Dog barks")

class Cat(Animal):
    def speak(self):  
        print("Cat meows")

print("Task 4: Method Overriding:")
animals = [Animal(), Dog(), Cat()]
for animal in animals:
    animal.speak()
print()
```

<h4>5. Create a class for operator overloading which adds two Point Objects where Point has x & y values <br> 
e.g. if  <br>
P1(x=10,y=20) <br> 
P2(x=12,y=15)  <br>
P3=P1+P2 => P3(x=22,y=35) </h4>

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            return Point(self.x + other.x, self.y + other.y)
        return NotImplemented

    def __str__(self):
        return f"Point(x={self.x}, y={self.y})"

p1 = Point(10, 20)
p2 = Point(12, 15)
p3 = p1 + p2

print("Task 5: Operator Overloading:")
print(f"P1: {p1}")
print(f"P2: {p2}")
print(f"P3 = P1 + P2: {p3}")
```
