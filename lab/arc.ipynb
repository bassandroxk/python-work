# data types and types checking 

```python
x = 204189.52
print(type(x))  # This will output: <class 'float'>
```

```python
x = "hello"
y = 4112
print(type(x))  # This will output: <class 'str'>
print(type(y))  # This will output: <class 'int'>
```

```python
x = "42"
y = int(x)
print(type(y))  # This will output: <class 'int'>
```

```python
x = {1,23, 51}
y = {"a", 4141}
print(type(x))  # This will output: <class 'set'>
print(type(y))  # This will output: <class 'set'>
```

```python
x = "hi"
y = len(x) if isinstance(x, str) else x
print(type(y))  # This will output: <class 'int'>
print(y)        # This will output: 2
```

```python
a = {1, 2, 3,4,5,6,7,8,9}
b = {5,6,7,8,9,10,11,12,1}
c = a.union(b)
print(type(c))  # This will output: <class 'set'>
print(c)
```

```python
a = {1414, 4115, 51, 51516, 1575167}
b = {51516, 1575167, 17171718, 19191919}
c = a.intersection(b)
print(c)
d = b.union(a)
print(d)
```

```python
import numpy as np
array = np.array([1, 2, 3, 4, 5])
print(type(array))
```

```python
array = np.array(["TRUE", 1, 3, 51, 61, 2726])
print(array)
print(type(array))
b = array
print(b)
```
