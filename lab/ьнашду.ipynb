## Dynamic Types in Python

### What are Dynamic Types?

**Dynamic typing** is a programming language feature where variable types are determined at runtime rather than at compile-time. In dynamically-typed languages like Python, you don't need to declare a variable's type before using it.

### Key Characteristics

| Feature | Description |
|---------|-------------|
| **Runtime Type Checking** | Types are checked during program execution |
| **Type Flexibility** | Variables can hold any data type and change types |
| **No Declaration Required** | Variables are assigned without type annotations |
| **Type Inference** | The interpreter infers the type from the assigned value |

### Examples

```python
# Python - dynamic typing
x = 5              # x is an int
x = "hello"        # x is now a string
x = [1, 2, 3]      # x is now a list
```

### Advantages vs Disadvantages

**‚úì Advantages:**
- Faster development and prototyping
- More concise and readable code
- Greater flexibility
- No verbose type declarations

**‚úó Disadvantages:**
- Runtime errors may occur
- Harder to maintain large codebases
- Performance overhead from type checking
- Limited IDE autocomplete support


```

## üî¢ Armstrong Number Checker

An **Armstrong number** (also called narcissistic number) is a number that equals the sum of its own digits each raised to the power of the number of digits.

**Example:** 153 = 1¬≥ + 5¬≥ + 3¬≥

````python
num = int(input("Enter a number: "))
power = len(str(num))

if num == sum(int(d) ** power for d in str(num)):
    print("Armstrong number")
else:
    print("Not an Armstrong number")





````
