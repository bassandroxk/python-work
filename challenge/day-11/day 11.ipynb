#### Q1. Take a list and return only prime numbers using function.

```
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def get_primes(lst):
    result = []
    for x in lst:
        if is_prime(x):
            result.append(x)
    return result

print(get_primes([1,2,3,4,5,6,7,8]))
```

#### 9 Take a string and print characters that appear more than once.

```

s = input("Enter a string: " )
freq = {}
for c in s:
    freq[c] = freq.get(c, 0) + 1
for ch, count in freq.items():
    if count > 1:
        print(f"{ch}: {count}")
```

#### Q8. Count how many elements are greater than average.

```
lst = [1, 2, 3, 4, 5]
avg = sum(lst) / len(lst)
count = sum(1 for x in lst if x > avg)
print("Number of elements greater than average:", count)
```

#### 9. Write a function to return multiple values (sum & average).

```
def sum_and_average(lst):
    total = sum(lst)
    average = total / len(lst)
    return total, average
lst = [int(x) for x in input("Enter numbers with space: " ).split()]
total, average = sum_and_average(lst)
print("Total =", total)
print("Average =", average)
```
