#### Q1. Print string with index using enumerate().

```python
s=input("Enter a string: ")
for i,c in enumerate(s):
    print(f"Index: {i}, Character: {c}")
```

#### Q3. Split a sentence into words.

```python
s = input("Enter a sentence: ")

words = s.split()

for w in words:
    print(w)
```
