#### Q1. Write a function to reverse a number.

```python
def reverse_no(n):
    rev=0
    while n>0:
        digit=n%10
        rev=rev*10+digit
        n//=10
    return rev

n=int(input("Enter a number: "))
print("Reverse =", reverse_no(n))
```

#### 7. Take a list and print sum of squares using loop.

```python
lst = [1, 2, 3, 4, 5]
total = 0
for x in lst:
    total += x * x
print("Sum of squares:", total)
```

<h4> 8. Create a dictionary of n persons where key is name and value is city.<br>   a. Display all names <br> b. Display all city names  <br>c. Display student names and cities of all students.<br>  d. Count the number of students in each city. </h4>

```python
stu={}
n=int(input("enter a how many students number:"))
for i in range(1,n+1):
    student=input("enter student name:")
    city=input("enter city name:")
    stu[student]=city
print(stu)
print("\nStudents name are")
for name in stu.keys():
    print(name)
print("\nCities are")
for city in stu.values():
    print (city)
print("\nStudent and their cities are")
for name,city in stu.items():
    print(name,"lives in", city)

city_count={}
for city in stu.values():
    if city in city_count:
        city_count[city]+=1
    else:
        city_count[city]=1

print("\nNumber of students in each city")
for city,count in city_count.items():
    print(city,"has" ,count,"students")

```

#### Q9. Find second largest in list (no built-in sort)

```python
lst=[10,20,5,15,25]
largest=second_largest=float('-inf')
for x in lst:
    if x>largest:
        second_largest=largest
        largest=x
    elif largest>x>second_largest:
        second_largest=x
print("Second largest:", second_largest)
```
