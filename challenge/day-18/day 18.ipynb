```python
t = (10, 20, 30, 40)
avg = sum(t) / len(t)
print("Average =", avg)
```

#### 7. Write a recursive function for Fibonacci.

```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

n=int(input("Enter a number: "))
fib(n)
```

#### 2 Check whether a number is an Armstrong number.

```python
n = int(input("Enter a number: "))
temp = n
s = 0
while temp > 0:
    digit = temp % 10
    s += digit ** 3
    temp //= 10
if s == n:
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is not an Armstrong number.")
```
