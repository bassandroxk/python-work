#### Q1. Check whether a number is Armstrong number.

```python
n=int(input("Enter a number: "))
temp=n
sum=0
while temp>0:
    digit=temp%10
    sum+=digit**3
    temp//=10
if sum==n:
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is not an Armstrong number.")
```

<h4> 2. Create 2 sets s1 and s2 of n fruits each by taking input from user and find: <br>   a. Fruits which are in both sets s1 and s2  <br> b. Fruits only in s1 but not in s2  <br> c. Count of all fruits from s1 and s2 </h4>

```python
n=int(input("enter a number: "))

s1=set()
s2=set()

print("Enter fruits for set1: ")
for i in range(n):
    fruit=input()
    s1.add(fruit)

print("Enter fruits for set2: ")
for i in range(n):
    fruit=input()
    s2.add(fruit)

print("Fruits in both s1 and s2:", s1 & s2)

print("Fruits only in s1:", s1-s2)

print("Total count of unique fruits:", len(s1|s2))
```

#### Q3. Count words in a sentence and store in dictionary.

```python
s=input("Enter a sentence: ")
words=s.split()
freq={}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)
```

#### Q4. Count uppercase, lowercase and digits.

```python
s=input("Enter a string: ")
upper=0
lower=0
digits=0
for c in s:
    if c.isupper():
        upper+=1
    elif c.islower():
        lower+=1
    elif c.isdigit():
        digits+=1
print(f"The number of uppercase letters in the string is: {upper}")
print(f"The number of lowercase letters in the string is: {lower}")
print(f"The number of digits in the string is: {digits}")
```
