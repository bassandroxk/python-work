#### 1. Write a function to find maximum of two numbers.

```
def maximum(a, b):
    if a > b:
        return a
    else:
        return b
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Maximum =", maximum(a, b))
```

#### Q3. Find largest number in list using function (no max).

```
def largest_list(lst):
    if not lst:
        return None
    max_val = lst[0]
    for x in lst:
        if x > max_val:
            max_val = x
    return max_val

x=[int(x) for x in input("Enter numbers with space: ").split()]
print("Largest number:", largest_list(x))
```

#### Q5. Check whether a number is prime.

```
n=int(input("Enter a number: "))
if n>1:
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not a prime number.")
            break
    else:
        print(f"{n} is a prime number.")
```

#### Q6. Count duplicate words in sentence

```

sentence = "This is a test sentence with some duplicate words. This sentence is a test."
words = sentence.split()
duplicate_words = []
for word in words:
    if words.count(word) > 1 and word not in duplicate_words:
        duplicate_words.append(word)
print("Duplicate words:", duplicate_words)
```

#### 9. Check if a number is a perfect number.

```

n = int(input("Enter a number: ") )
divisors = [i for i in range(1, n) if n % i == 0]
if sum(divisors) == n:
    print(f"{n} is a perfect number.")
else:
    print(f"{n} is not a perfect number.")
```
