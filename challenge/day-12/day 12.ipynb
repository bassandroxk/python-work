#### 1. Check whether a number is prime.

```python

n = int(input("Enter a number: ") )
if n > 1:
    for i in range(2, n):
        if n % i == 0:
            print(f"{n} is not a prime number.")
            break
    else:
        print(f"{n} is a prime number.")
else:
    print(f"{n} is not a prime number.")
```

#### Q2. Print pyramid:</h4>

```python
n = 4
for i in range(1, n+1):
    print(" "*(n-i) + "*"*(2*i-1))
```

#### 3. Student Marks System (Dictionary)Take 3 students marks and print average marks.

```python
marks = {}
for i in range(3):
    name = input("Enter student name: ")
    mark = float(input("Enter marks: "))
    marks[name] = mark

avg = sum(marks.values()) / len(marks)
print("Average marks: %.2f" %avg)
```

#### Q4. Take a number and print all its factors.

```python
n=int(input("Enter a number: "))

for i in range(1, n+1):
    if n % i == 0:
        print(i, end=" ")
```

#### Q5. Flatten a 2D list.

```python
lst=[[1,2],[3,4],[5,6]]
flat=[x for sublist in lst for x in sublist]
print(flat)
```

#### 6. Take a string and print frequency of vowels using dictionary.

```python
s = input("Enter a string: " ).lower()
freq = {"a":0, "e":0, "i":0, "o":0, "u":0}
for ch in s:
    if ch in freq:
        freq[ch] += 1
print(freq)
```

#### 4 Find sum of elements in a list.

```python

numbers = [10, 20, 30, 40]
total = 0
for x in numbers:
    total += x
print("Total:", total)
```
